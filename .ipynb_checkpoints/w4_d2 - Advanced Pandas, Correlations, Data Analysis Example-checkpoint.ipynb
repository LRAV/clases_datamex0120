```python
animals=pd.read_csv('data/animals.csv')
animals.head()
```

```python
animals['bodywtkg']=animals.bodywt*0.4536
animals.head()
```

```python
animals['wtratio']=animals.bodywt/animals.brainwt
animals.head()
```

```python
animals['zerocheck']=np.where(animals.brainwt!=0,
                              animals.bodywt/animals.brainwt, 0)
animals.head()
```

```python
# lookup

seasons=pd.DataFrame(np.random.random((10, 4)), 
                     columns=['winter', 'spring', 'summer', 'autumn'])
seasons
```

```python
look=['summer', 'winter', 'spring', 
      'winter', 'summer', 'autumn',
      'summer','summer', 'winter', 'spring']

seasons.lookup(seasons.index, look)
```

```python
# get

seasons.get(seasons.winter>0.5)
```

```python
# index y reindex

url='https://archive.ics.uci.edu/ml/machine-learning-databases/auto-mpg/auto-mpg.data'

cols=['mpg', 'cylinders', 'displacement', 'horsepower', 
      'weight', 'acceleration', 'model_year', 'origin', 'car_name']
```

```python
auto=pd.read_csv(url, sep='\\s+', names=cols)
auto.head()
```

```python
auto.index
```

```python
auto.index=pd.RangeIndex(0, 398*2, 2)
auto.head()
```

```python
auto.index=range(len(auto))
auto.head()
```

```python
auto.index=auto.car_name
auto.head()
```

```python
auto.set_index('mpg', inplace=True)
auto.head()
```

```python
auto.describe()#.loc['mean']
```
